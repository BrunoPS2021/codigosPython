```python
import pyautogui as posicaoAbreArquivos
import pyperclip

posicaoAbrirGoogle.click(x=120, y=547)

posicaoAbrirGoogle.sleep(2)
posicaoAbreArquivos.hotkey('win','r')

posicaoAbreArquivos.sleep(2)

posicaoAbreArquivos.typewrite('notepad')

posicaoAbrirGoogle.sleep(2)

posicaoAbrirGoogle.press('enter')

posicaoAbrirGoogle.sleep(3)

pyperclip.copy('Abrimos o notepad com um robô ou script')
posicaoAbrirGoogle.hotkey("ctrl","v")

fecharJanelaNotepad = posicaoAbrirGoogle.getActiveWindow()

fecharJanelaNotepad.close()

posicaoAbrirGoogle.sleep(2)

posicaoAbrirGoogle.press('tab')

posicaoAbrirGoogle.sleep(2)

posicaoAbrirGoogle.press('enter')
```

```python
import pyperclip
import pyautogui
import pyautogui as escolha_opcao

opcao = pyautogui.confirm('Clique no botão desejado', buttons = ['Excel','Word','Notepad'])

if opcao == "Excel":
    escolha_opcao.click(x=120, y=547)
    escolha_opcao.sleep(2)
    escolha_opcao.hotkey('win','r')
    escolha_opcao.sleep(2)
    escolha_opcao.typewrite('excel')
    escolha_opcao.sleep(2)
    escolha_opcao.press('enter')
    escolha_opcao.sleep(7)
    #print(escolha_opcao.position())
    escolha_opcao.click(x=314, y=227)
    escolha_opcao.sleep(7)
    pyperclip.copy('Escolhi abrir o excel')
    escolha_opcao.hotkey('ctrl','v')
elif opcao == "Word":
    escolha_opcao.click(x=120, y=547)
    escolha_opcao.sleep(2)
    escolha_opcao.hotkey('win','r')
    escolha_opcao.sleep(2)
    escolha_opcao.typewrite('winword')
    escolha_opcao.sleep(2)
    escolha_opcao.press('enter')
    escolha_opcao.sleep(7)
    print(escolha_opcao.position())
    escolha_opcao.click(x=314, y=227)
    escolha_opcao.sleep(7)
    pyperclip.copy('Escolhi abrir o word')
    escolha_opcao.hotkey('ctrl','v')
elif opcao == "Notepad":
    escolha_opcao.click(x=120, y=547)
    escolha_opcao.sleep(2)
    escolha_opcao.hotkey('win','r')
    escolha_opcao.sleep(2)
    escolha_opcao.typewrite('notepad')
    escolha_opcao.sleep(2)
    escolha_opcao.press('enter')
    escolha_opcao.sleep(7)
    pyperclip.copy('Escolhi abrir o notepad')
    escolha_opcao.hotkey('ctrl','v')
```
