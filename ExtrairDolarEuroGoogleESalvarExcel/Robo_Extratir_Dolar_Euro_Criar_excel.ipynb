```python
#Importando o selenium para trabalhar com as páginas da web
from selenium import webdriver as opcoes_selenium_aula
from selenium.webdriver.common.keys import Keys

#Importando a biblioteca do pyautogui para trabalhar o tempo e teclas
import pyautogui as tempoPausaComputador

#Usando o pyautogi para controlar as teclas do teclado
import pyautogui as teclasAtalhoTeclado

#Usando o By para trabalhar com as atualizações mais recentes
from selenium.webdriver.common.by import By

#Passando autorização ao acesso as configurações do Chrome
meuNavegador = opcoes_selenium_aula.Chrome()
meuNavegador.get('https://www.google.com.br/')

#Aguarda 5 segundos para o computador processar as informações
tempoPausaComputador.sleep(5)

#Procurar pelo element ID e quando encontrar escreve "Dolar hoje"
meuNavegador.find_element(By.NAME,'q').send_keys('Dolar hoje')

#Aguarda 5 segundos para o computador processar as informações
tempoPausaComputador.sleep(5)

#Retornar para o campo name q
#Faz a busca do valor que esta digitado no campo NAME q
meuNavegador.find_element(By.NAME,'q').send_keys(Keys.RETURN)

#Aguarda 5 segundos para o computador processar as informações
tempoPausaComputador.sleep(5)

#No resultado da pesquisa pegar XPATH e no meio pegar o primeiro elemento da lista
valorDolarPeloGoogle = meuNavegador.find_elements(By.XPATH,'//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]')[0].text

#Aguarda 5 segundos para o computador processar as informações
tempoPausaComputador.sleep(5)

print(valorDolarPeloGoogle)

#-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

#Aguarda 5 segundos para o computador processar as informações
tempoPausaComputador.sleep(5)

meuNavegador.find_element(By.NAME,'q').send_keys('')

#Aguarda 5 segundos para o computador processar as informações
tempoPausaComputador.sleep(5)

#Apertar a tecla tab
teclasAtalhoTeclado.press('tab')

#Aguarda 5 segundos para o computador processar as informações
tempoPausaComputador.sleep(5)

#Apertar a tecla enter
#Limpar o campo de pesquisa
teclasAtalhoTeclado.press('enter')

#Procurar pelo element ID e quando encontrar escreve "Euro hoje"
meuNavegador.find_element(By.NAME,'q').send_keys('Euro hoje')

#Aguarda 5 segundos para o computador processar as informações
tempoPausaComputador.sleep(5)

#Retornar para o campo name q
#Faz a busca do valor que esta digitado no campo NAME q
meuNavegador.find_element(By.NAME,'q').send_keys(Keys.RETURN)

#Aguarda 5 segundos para o computador processar as informações
tempoPausaComputador.sleep(5)

#No resultado da pesquisa pegar XPATH e no meio pegar o primeiro elemento da lista
valorEuroPeloGoogle = meuNavegador.find_elements(By.XPATH,'//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]')[0].text

#Aguarda 5 segundos para o computador processar as informações
tempoPausaComputador.sleep(5)

print(valorEuroPeloGoogle)

#------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

import xlsxwriter
import os

nomeCaminhoArquivo = "C:\\Python\\ExtrairDolarEuroGoogleESalvarExcel\\Dolar e Euro Google.xlsx"
planilhaCriada = xlsxwriter.Workbook(nomeCaminhoArquivo)
planilha1 = planilhaCriada.add_worksheet()

#Escrevendo nas celulas
planilha1.write("A1", "Dolar")
planilha1.write("B1", "Euro")
planilha1.write("A2", valorDolarPeloGoogle)
planilha1.write("B2", valorEuroPeloGoogle)

#Substitui , por .
valorDolarPeloGoogle = valorDolarPeloGoogle.replace(',','.')
valorEuroPeloGoogle = valorEuroPeloGoogle.replace(',','.')

#Converter string em float
valorDolarTipoFloat = float(valorDolarPeloGoogle)
valorEuroTipoFloat = float(valorEuroPeloGoogle)

planilha1.write("A3", valorDolarTipoFloat)
planilha1.write("B3", valorEuroTipoFloat)

#Fechar o arquivo do Excel que está em segundo plano
planilhaCriada.close()

#Abrir arquivo
os.startfile(nomeCaminhoArquivo)
```
