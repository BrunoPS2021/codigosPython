```python
import tkinter as tk
from tkinter import messagebox

def gerar_tabuada():
    try:
        numero = int(entrada_numero.get().strip())
    except ValueError:
        messagebox.showwarning("Entrada inválida", "Por favor, insira um número inteiro válido")
        return
    
    tabuada_resultado = []

    for i in range(1,11):
        tabuada_resultado.append(f"{numero} X {i} = {numero*i}")
        entrada_tabuada.delete("1.0",tk.END)
        entrada_tabuada.insert(tk.END, "\n".join(tabuada_resultado))



janela_principal = tk.Tk()
janela_principal.title("Tabuada")
janela_principal.config(bg="white")

largura_janela = 300
altura_janela = 400

largura_tela = janela_principal.winfo_screenwidth()
altura_tela = janela_principal.winfo_screenheight()

posicao_x = (largura_tela // 2) - (largura_janela // 2)
posicao_y = (altura_tela // 2) - (altura_janela // 2)

janela_principal.geometry(f"{largura_janela}x{altura_janela}+{posicao_x}+{posicao_y}")

frame_numero = tk.Frame(janela_principal, bg="white")
frame_numero.pack(pady=10)

label_numero = tk.Label(frame_numero,
                        text="Número: ",
                        bg="white",
                        font=('Helvetica',12))
label_numero.grid(row=0, column=0, padx=5, pady=5)

entrada_numero = tk.Entry(frame_numero,width=20,font=('Helvetica',12))
entrada_numero.grid(row=0, column=1, padx=5, pady=5)

botao_gerar = tk.Button(janela_principal,
                        text="Gerar Tabuada",
                        command=gerar_tabuada,
                        font=('Helvetica',12))
botao_gerar.pack(pady=10)

entrada_tabuada = tk.Text(janela_principal,
                         width=25,
                         height=10,
                         font=('Helvetica',14))
entrada_tabuada.pack(pady=10)

janela_principal.mainloop()
```
