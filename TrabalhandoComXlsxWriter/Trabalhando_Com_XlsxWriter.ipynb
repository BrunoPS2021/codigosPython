```python
import xlsxwriter as opcoesDoXlswriter
import os

nomeCaminhoArquivo = 'Formulas.xlsx'
minhaPlanilha = opcoesDoXlswriter.Workbook(nomeCaminhoArquivo)
sheetDados = minhaPlanilha.add_worksheet('Dados')

sheetDados.write('A1', 'Número 1')
sheetDados.write('B1', 'Número 2')
sheetDados.write('C1', 'Fórmula')

sheetDados.write('A2', 10)
sheetDados.write('A3', 6)
sheetDados.write('A4', 8)
sheetDados.write('A5', 6)
sheetDados.write('A8', 'Chloé')

sheetDados.write('B2', 7)
sheetDados.write('B3', 5)
sheetDados.write('B4', 3)
sheetDados.write('B5', 1)
sheetDados.write('B8', 'Silva')

sheetDados.write_formula('C2', '=A2+B2')
sheetDados.write_formula('C3', '=A3-B3')
sheetDados.write_formula('C4', '=A4*B4')
sheetDados.write_formula('C5', '=A5/B5')
sheetDados.write_formula('C8', '=CONCATENATE(A8," ",B8)')

minhaPlanilha.close()

os.startfile(nomeCaminhoArquivo)
```

```python
import xlsxwriter as opcoesDoXlswriter
import os

nomeCaminhoArquivo = 'MergeCells.xlsx'
minhaPlanilha = opcoesDoXlswriter.Workbook(nomeCaminhoArquivo)
sheet = minhaPlanilha.add_worksheet()

add_merge_celulas = workbook.add_format({
    'bold': True,
    'border': 6,
    'align': 'center',
    'valign': 'vcenter',
    'size': 30,
    'fg_color': 'blue',
    'font_color': 'white'
})

sheet.merge_range('B3:I5', 'Aula Merge Cells', add_merge_celulas)

minhaPlanilha.close()

os.startfile(nomeCaminhoArquivo)
```

```python
import xlsxwriter as opcoesDoXlswriter
import os

nomeCaminhoArquivo = 'FormatacaoCondicional.xlsx'
minhaPlanilha = opcoesDoXlswriter.Workbook(nomeCaminhoArquivo)
sheet = minhaPlanilha.add_worksheet('Dados')

formatoMaior = minhaPlanilha.add_format({
    'bg_color': 'green',
    'font_color': 'white'
})

formatoMenor = minhaPlanilha.add_format({
    'bg_color': 'red',
    'font_color': 'white'
})

inserirDados = [
    ['Coluna 1', 'Coluna 2', 'Coluna 3', 'Coluna 4'],
    [34, 50, 12, 34],
    [23, 32, 76, 51],
    [43, 29, 34, 12],
    [29, 58, 73, 19]
]

sheet.write('A1', 'Células com os valores >= 50 estão em verde e < 50 estão em vermelho')

for linha, range in enumerate(inserirDados):
    sheet.write_row(linha + 2, 1, range) #linha 3, coluna 2

sheet.conditional_format('B4:E7', {
    'type': 'cell',
    'criteria': '>=',
    'value': 50,
    'format': formatoMaior
    })

sheet.conditional_format('B4:E7', {
    'type': 'cell',
    'criteria': '<',
    'value': 50,
    'format': formatoMenor
    })

minhaPlanilha.close()

os.startfile(nomeCaminhoArquivo)
```

```python
import xlsxwriter as opcoesDoXlswriter
import os

nomeCaminhoArquivo = 'FormatacaoCondicionalIcones.xlsx'
minhaPlanilha = opcoesDoXlswriter.Workbook(nomeCaminhoArquivo)
sheet = minhaPlanilha.add_worksheet('Dados')

inserirDados = [
    ['Coluna 1', 'Coluna 2', 'Coluna 3', 'Coluna 4'],
    [34, 50, 12, 34],
    [23, 32, 76, 51],
    [43, 29, 34, 12],
    [29, 58, 73, 19],
    [18, 30, 45, 12],
]

sheet.write('A1', 'Exemplo condicional com icones')

for linha, range in enumerate(inserirDados):
    sheet.write_row(linha + 2, 1, range) #linha 3, coluna 2

sheet.conditional_format('B4:E4', {
    'type': 'icon_set',
    'icon_style': '3_traffic_lights'
    })

sheet.conditional_format('B5:E5', {
    'type': 'icon_set',
    'icon_style': '3_traffic_lights',
    'reverse_icons': True
    })

sheet.conditional_format('B6:E6', {
    'type': 'icon_set',
    'icon_style': '3_arrows'
    })

sheet.conditional_format('B7:E7', {
    'type': 'icon_set',
    'icon_style': '4_arrows'
    })

sheet.conditional_format('B8:E8', {
    'type': 'icon_set',
    'icon_style': '5_ratings'
    })



minhaPlanilha.close()

os.startfile(nomeCaminhoArquivo)
```

```python
import xlsxwriter as opcoesDoXlswriter
import os

nomeCaminhoArquivo = 'Grafico.xlsx'
minhaPlanilha = opcoesDoXlswriter.Workbook(nomeCaminhoArquivo)
sheet = minhaPlanilha.add_worksheet('Resumo')

linhaNegrito = minhaPlanilha.add_format({'bold': 1})

titulos = ['Vendedores', 'Total Vendas']
dadosTabela = [
    ['Ana', 'Pedro', 'Bruno', 'Deyse', 'Chloé', 'Miguel'],
    [400, 300, 89, 34, 350, 120]
]

sheet.write_row('A1', titulos, linhaNegrito)
sheet.write_column('A2', dadosTabela[0])
sheet.write_column('B2', dadosTabela[1])

graficosColunas = minhaPlanilha.add_chart({'type': 'column'})

graficosColunas.add_series({
    'name': '=Resumo!$B$1',
    'categories': '=Resumo!$A$2:$A$7',
    'values': '=Resumo!$B$2:$B$7',
})

graficosColunas.set_title({
    'name': 'Gráfico Total de vendas'
})

graficosColunas.set_x_axis({
    'name': 'Vendedores'
})

graficosColunas.set_y_axis({
    'name': 'Vendas'
})

graficosColunas.set_style(11)

sheet.insert_chart('D2',graficosColunas, {'x_offset': 25, 'y_offset': 10})


####################################################################################


graficosEmpilhado = minhaPlanilha.add_chart({'type': 'area', 'subtype': 'stacked'})

graficosEmpilhado.add_series({
    'name': '=Resumo!$B$1',
    'categories': '=Resumo!$A$2:$A$7',
    'values': '=Resumo!$B$2:$B$7',
})

graficosEmpilhado.set_title({
    'name': 'Gráfico Empilhado'
})

graficosEmpilhado.set_x_axis({
    'name': 'Funcionários'
})

graficosEmpilhado.set_y_axis({
    'name': 'Vendas'
})

graficosEmpilhado.set_style(12)

sheet.insert_chart('M2',graficosEmpilhado, {'x_offset': 25, 'y_offset': 10})

minhaPlanilha.close()

os.startfile(nomeCaminhoArquivo)
```
