```python
import tkinter as tk
from tkinter import messagebox

def calcular_aumento():
    try:
        salario_atual = float(entrada_salario_atual.get())
        salario_novo  = float(entrada_salario_novo.get())

        if salario_atual <= 0 or salario_novo <= 0:
            raise ValueError('Os sal치rios devem ser maiores que zero.')
        
        valor_aumento = salario_novo - salario_atual 
        percentual = valor_aumento / salario_atual * 100

        entrada_valor_aumento.config(state=tk.NORMAL)
        entrada_valor_aumento.delete(0,tk.END)
        entrada_valor_aumento.insert(0,f"R$ {valor_aumento:.2f}")
        entrada_valor_aumento.config(state=tk.DISABLED)
        
        entrada_percentual_aumento.config(state=tk.NORMAL)
        entrada_percentual_aumento.delete(0,tk.END)
        entrada_percentual_aumento.insert(0,f"{percentual:.2f} %")
        entrada_percentual_aumento.config(state=tk.DISABLED)

        print("...")
    except ValueError as ve:
        messagebox.showerror('Erro', f"Entrada inv치lida: {ve}")

    except Exception as e:
        messagebox.showerror('Erro', f"Erro inesperado: {e}")

janela_principal = tk.Tk()
janela_principal.title("Calculador de Aumento Salarial")
janela_principal.config(bg="white")

largura_janela = 400
altura_janela = 300

largura_tela = janela_principal.winfo_screenwidth()
altura_tela = janela_principal.winfo_screenheight()

posicao_x = (largura_tela // 2) - (largura_janela // 2)
posicao_y = (altura_tela // 2) - (altura_janela // 2)

janela_principal.geometry(f"{largura_janela}x{altura_janela}+{posicao_x}+{posicao_y}")

frame_entrada = tk.Frame(janela_principal, bg="white")
frame_entrada.pack(pady=20)

label_salario_atual = tk.Label(frame_entrada,
                        text="Sal치rio Atual (R$): ",
                        bg="white",
                        font=('Helvetical', 12))
label_salario_atual.grid(row=0, column=0,padx=5,pady=5)

entrada_salario_atual = tk.Entry(frame_entrada,
                          width=20,
                          font=('Helvetical', 12))
entrada_salario_atual.grid(row=0,column=1,padx=5,pady=5)

label_salario_novo = tk.Label(frame_entrada,
                        text="Novo Sal치rio (R$): ",
                        bg="white",
                        font=('Helvetical', 12))
label_salario_novo.grid(row=1, column=0,padx=5,pady=5)

entrada_salario_novo = tk.Entry(frame_entrada,
                          width=20,
                          font=('Helvetical', 12))
entrada_salario_novo.grid(row=1,column=1,padx=5,pady=5)

botao_calcular_aumento = tk.Button(janela_principal,
                          text="Calcular Aumento",
                          command=calcular_aumento,
                          font=('Helvetical', 12))
botao_calcular_aumento.pack(pady=20)

frame_resultado = tk.Frame(janela_principal, bg="white")
frame_resultado.pack(pady=20)

label_valor_aumento = tk.Label(frame_resultado,
                        text="Valor do Aumento: ",
                        bg="white",
                        font=('Helvetical', 12))
label_valor_aumento.grid(row=0, column=0,padx=5,pady=5)

entrada_valor_aumento = tk.Entry(frame_resultado,
                          width=20,
                          state=tk.DISABLED,
                          font=('Helvetical', 12))
entrada_valor_aumento.grid(row=0,column=1,padx=5,pady=5)

label_percentual_aumento = tk.Label(frame_resultado,
                        text="Percentual do Aumento: ",
                        bg="white",                        
                        font=('Helvetical', 12))
label_percentual_aumento.grid(row=1, column=0,padx=5,pady=5)

entrada_percentual_aumento = tk.Entry(frame_resultado,
                          width=20,
                          state=tk.DISABLED,
                          font=('Helvetical', 12))
entrada_percentual_aumento.grid(row=1,column=1,padx=5,pady=5)

janela_principal.mainloop()
```
