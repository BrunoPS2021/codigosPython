```python
import tkinter as tk
import pandas as pd

def carregar_dados_excel(filtro=None):
    df = pd.read_excel("tabela_exemplo.xlsx")
    
    colunas = list(df.columns)

    if filtro:
        df = df[df.apply(lambda row: filtro.lower() in row.astype(str).str.lower().to_string(), axis=1)]

    for j, coluna in enumerate(colunas):
        cabecalho[j].config(text=coluna)

    for i in range(len(tabela)):
        for j in range(len(tabela[i])):
            if i < len(df):
                tabela[i][j].config(text=df.iloc[i, j])
            else:
                tabela[i][j].config(text="")

def aplicar_filtro(event):
    filtro = entrada_filtro.get()

    carregar_dados_excel(filtro=filtro)

def exportar_para_excel():
    dados = []
    for i in range(len(tabela)):
        linha = []
        for j in range(len(tabela[i])):
            valor = tabela[i][j].cget("text")
            linha.append(valor)

        dados.append(linha)

    df = pd.DataFrame(dados,columns=[cabecalho[j].cget("text") for j in range(len(cabecalho))])

    df.to_excel("Tabela_Exportada.xlsx",index=False)
    
    print("Dados Exportados para a Tabela_Exportada.xlsx")

def destacar_celula(event):
    linha = event.widget.grid_info()['row'] - 3
    coluna = event.widget.grid_info()['column']

    if 0 <= linha < len(tabela) and 0 <= coluna < len(tabela[0]):
        for i in range(len(tabela)):
            for j in range(len(tabela[i])):
                tabela[i][j].config(bg="white",
                                    fg="black",
                                    font=("Arial",10))
                
        tabela[linha][coluna].config(bg="yellow",
                                     fg="black",
                                     font=("Arial",14, "bold"))
        
        for j in range(len(tabela[linha])):
            if j != coluna:
                tabela[linha][j].config(bg="lightgray")

        for i in range(len(tabela)):
            if i != linha:
                tabela[i][coluna].config(bg="lightgray")

def restaurar_celula(event):
    for i in range(len(tabela)):
        for j in range(len(tabela[i])):
            tabela[i][j].config(bg="white",
                                fg="black",
                                font=("Arial",10))

janela = tk.Tk()
janela.title("Tabela Interativa com Dados Excel")

titulo = tk.Label(janela,
                  text="Tabela de Infromações",
                  font=("Arial",16,"bold"))
titulo.grid(row=0,column=0,columnspan=5, pady=10)

entrada_filtro = tk.Entry(janela, font=("Arial",12))
entrada_filtro.grid(row=1,column=0,padx=5,pady=5,sticky="ew")
entrada_filtro.bind("<KeyRelease>",aplicar_filtro)

btn_exportar = tk.Button(janela,text="Exportar para Excel", 
                         command=exportar_para_excel,
                         font=("Arial",12))
btn_exportar.grid(row=1,column=4,padx=5,pady=5)

df = pd.read_excel("tabela_exemplo.xlsx")

colunas = len(df.columns)
linhas = len(df)

cabecalho = []

for j in range(colunas):
    celula_cabecalho = tk.Label(janela,
                                text='',
                                bg="lightblue",
                                fg="black",
                                width=12, height=2,
                                borderwidth=1, relief="solid",
                                font=("Arial",12,"bold"))
    celula_cabecalho.grid(row=2, column=j, sticky="nsew")
    cabecalho.append(celula_cabecalho)

tabela = []

for i in range(linhas):
    linha = []
    for j in range(colunas):
        celula = tk.Label(janela,
                          text="",
                          bg="white",
                          fg="black",
                          width=12,height=3,
                          borderwidth=1, relief="solid",
                          font=("Arial",10))
        celula.grid(row=i+3,column=j,sticky="nsew")
        celula.bind("<Enter>",destacar_celula)
        celula.bind("<Leave>",restaurar_celula)

        linha.append(celula)

    tabela.append(linha)

for j in range(colunas):
    janela.grid_columnconfigure(j, weight=1)

for i in range(linhas):
    janela.grid_rowconfigure(i+3, weight=1)

carregar_dados_excel()

janela.mainloop()
```
